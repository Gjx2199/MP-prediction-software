```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel('Model 2.xlsx')#引入数据，数据和代码在一个文件夹
data=pd.DataFrame(data)#与上行意思相同
X = data.values[:,0:12].reshape(-1,12)#重新选中矩阵
Y = data.values[:,12:14].reshape(-1,2)#定义后面三个为y
print(X[0])
print(X.shape)#x有多少列
print(Y[0])
print(Y.shape)
data.describe()#将数据以表格呈现
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
#mm_X=StandardScaler()
#mm_Y=StandardScaler()
#X=mm_X.fit_transform(X)
#Y=mm_Y.fit_transform(Y)#数据归一化

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=42)#选择要不要归一化，一般归一化结果较好
```

```python
import numpy as np
import warnings
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_error  # 导入 mean_absolute_error 函数

# 设置忽略警告
warnings.filterwarnings('ignore')

# 假设之前得到的最佳超参数
best_n_estimators = 256
best_max_depth = 37
best_min_samples_split = 2
best_min_samples_leaf = 2
best_max_features = 'sqrt'
best_bootstrap = False

# 初始化随机森林回归模型，并设置最佳超参数
model = RandomForestRegressor(n_estimators=best_n_estimators,
                              max_depth=best_max_depth,
                              min_samples_split=best_min_samples_split,
                              min_samples_leaf=best_min_samples_leaf,
                              max_features=best_max_features,
                              bootstrap=best_bootstrap,
                              random_state=42)

# 使用之前划分好的训练集和测试集进行训练和预测
model.fit(X_train, y_train)
y_train_pre = model.predict(X_train)
y_test_pre = model.predict(X_test)

# 输出模型评估指标
r1 = r2_score(y_train, y_train_pre, multioutput='raw_values')
r2 = r2_score(y_test, y_test_pre, multioutput='raw_values')

trMSE = mean_squared_error(y_train, y_train_pre, multioutput='raw_values')
trRMSE = trMSE**0.5
teMSE = mean_squared_error(y_test, y_test_pre, multioutput='raw_values')
teRMSE = teMSE**0.5
# 计算训练集和测试集的 MAE
train_MAE = mean_absolute_error(y_train, y_train_pre, multioutput='raw_values')
test_MAE = mean_absolute_error(y_test, y_test_pre, multioutput='raw_values')

print('Train R2:', np.around(r1, 4))
print('Test R2:', np.around(r2, 4))
print('Average Train R2:', np.around(np.mean(r1, axis=0), 4))
print('Average Test R2:', np.around(np.mean(r2, axis=0), 4))

print('Train RMSE:', np.around(trRMSE, 4))
print('Test RMSE:', np.around(teRMSE, 4))
print('Average Train RMSE:', np.around(np.mean(trRMSE, axis=0), 4))
print('Average Test RMSE:', np.around(np.mean(teRMSE, axis=0), 4))

print('Train MAE:', np.around(train_MAE, 4))
print('Test MAE:', np.around(test_MAE, 4))
print('Average Train MAE:', np.around(np.mean(train_MAE, axis=0), 4))
print('Average Test MAE:', np.around(np.mean(test_MAE, axis=0), 4))


# 如果需要保存预测结果
data = {'y_train_pre': y_train_pre, 'y_test_pre': y_test_pre}
```
